# Estrategia de Cadena de Gimnasios Model Fitness

## Introducción
En la industria del fitness, captar nuevos clientes suele requerir una inversión considerable en publicidad, promociones y programas de afiliación. Sin embargo, mantener a los clientes actuales resulta significativamente más rentable que adquirir nuevos usuarios. Por esta razón, comprender las causas que llevan a un cliente a abandonar un gimnasio se ha convertido en una prioridad estratégica para muchas organizaciones.

Model Fitness, una cadena de gimnasios, enfrenta el desafío de reducir la pérdida de clientes (churn) y mejorar la retención de sus miembros. Para ello, resulta fundamental identificar los factores que influyen en la cancelación de membresías y detectar patrones de comportamiento que permitan anticipar el riesgo de abandono antes de que ocurra.

En este proyecto se emplearán técnicas de análisis exploratorio de datos, aprendizaje automático y segmentación de clientes para responder tres preguntas clave:

* ¿Qué características tienen los clientes con mayor probabilidad de abandonar el gimnasio?
* ¿Es posible predecir el churn antes de que ocurra?
* ¿Existen grupos de clientes con comportamientos diferenciados que requieran estrategias específicas de retención?

A partir de este análisis, se desarrollarán modelos predictivos y segmentos de clientes que permitan transformar los datos en acciones concretas para fortalecer la fidelización, optimizar los recursos de marketing y aumentar el valor de vida de los clientes.

### Histogramas de barras y distribuciones de características para aquellas personas que se fueron (cancelación) y para las que se quedaron.

```python
# Diferenciar los grupos de quienes se quedaron y cancelaron
stay = gym[gym['churn'] == 0]
churn = gym[gym['churn'] == 1]
```

```python
# Creación de histogramas
for feature in columns:
    plt.figure(figsize = (6,6))
    
    sns.histplot(stay[feature], kde=True, color='blue', label='Stay')
    sns.histplot(churn[feature], kde=True, color='red', label='Churn')

    plt.title(f'Distribución de {feature}')
    plt.legend()
    plt.show()
```

#### Comentario
Al comparar las distribuciones de las variables entre clientes que permanecen y quienes cancelan su membresía, se observan diferencias importantes.

* El género no parece tener una relación clara con la cancelación, ya que las distribuciones son bastante similares entre ambos grupos.
* Los clientes que viven o trabajan cerca del gimnasio (__near_location__) presentan una mayor permanencia. La cercanía del gimnasio a las viviendas o lugares de trabajo parece facilitar la asistencia y reducir el abandono.
* Las personas pertenecientes a empresas asociadas (__partner__) y quienes ingresaron mediante promociones con amigos (__promo_friends__) muestran una mayor tendencia a mantenerse en el gimnasio. Esto podría indicar que los beneficios o el componente social aumentan el compromiso.
* Los clientes con contratos más largos (__contract_period__) cancelan menos. Se observa que gran parte de quienes abandonan tienen contratos cortos, especialmente de un mes.
* La participación en clases grupales (__group_visits__) parece estar relacionada con una menor tasa de abandono, posiblemente por una mayor integración y compromiso con el gimnasio.
* Respecto a la edad, los clientes que permanecen tienden a concentrarse en edades ligeramente mayores, mientras que quienes cancelan presentan una distribución más hacia edades menores.
* Los usuarios que permanecen gastan más dinero en servicios adicionales (__avg_additional_charges_total__), lo que podría indicar un mayor nivel de involucramiento con el gimnasio.
* La variable __month_to_end_contract__ muestra una diferencia muy marcada: quienes cancelan suelen tener pocos meses restantes de contrato, mientras que quienes permanecen tienen contratos con mayor duración pendiente. Esto puede ser debido a que los clientes que permanecen en el gimnasio tienen un mayor costo de oportunidad al ser un extenso periodo restante, es decir, perderían mucho más al suspender el servicio al restar una cantidad significativa de meses.
* La antigüedad (__lifetime__) también presenta un patrón claro: los clientes nuevos tienen mayor probabilidad de abandonar, mientras que quienes llevan más tiempo suelen permanecer.
* Finalmente, las variables de frecuencia de asistencia (__avg_class_frequency_total__ y __avg_class_frequency_current_month__) muestran una de las diferencias más evidentes. Los clientes que permanecen visitan el gimnasio con mucha más frecuencia que quienes terminan cancelando.

## Predicción del Abandono de Clientes mediante Machine Learning
Tras identificar patrones de comportamiento y factores asociados a la cancelación de membresías, el siguiente paso consiste en desarrollar un modelo predictivo capaz de anticipar qué clientes presentan un mayor riesgo de abandono.

La capacidad de identificar clientes en riesgo antes de que cancelen su membresía representa una ventaja estratégica para Model Fitness, ya que permite implementar acciones preventivas de fidelización, optimizar recursos de marketing y reducir las pérdidas asociadas a la rotación de clientes.

Para ello, se construirá un modelo de clasificación binaria cuyo objetivo será predecir si un cliente abandonará o no el gimnasio durante el mes siguiente. El modelo aprenderá a partir de patrones históricos observados en los datos y utilizará variables relacionadas con el comportamiento, la actividad y las características de los usuarios para estimar la probabilidad de cancelación.

El desempeño de los modelos será evaluado mediante métricas de clasificación, permitiendo seleccionar la alternativa con mayor capacidad predictiva para apoyar futuras estrategias de retención.

**Pregunta de Negocio**

¿Es posible identificar con anticipación a los clientes con mayor riesgo de cancelar su membresía?

### Modelo de clasificación binaria para clientes donde la característica objetivo es la marcha del usuario o la usuaria el mes siguiente.

```python
# Primero, debemos separar variables y objetivo, es decir, x e y

# Características
X = gym.drop('churn', axis = 1)

# Variable Objetivo -> 'churn'
y = gym['churn']
```

### Construcción del Entorno de Entrenamiento
Con el fin de evaluar el desempeño de los algoritmos de clasificación, los datos serán separados en conjuntos de entrenamiento y prueba. Esta metodología permite medir la capacidad predictiva de los modelos sobre observaciones que no fueron utilizadas durante el proceso de aprendizaje.

```python
# Librería para hacer uso de la Función
from sklearn.model_selection import train_test_split
```

```python
# Datos de conjuntos de entrenamiento y validación
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.2,
    random_state = 0
)
```

```python
# Estandarizar datos
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)
```

### Entrenamiento y Comparación de Modelos Predictivos
Después de preparar los datos y definir la variable objetivo, el siguiente paso consiste en entrenar modelos de Machine Learning capaces de identificar clientes con riesgo de abandono.

Para este análisis se utilizarán dos algoritmos de clasificación ampliamente utilizados:

* **Regresión Logística**, que servirá como modelo base gracias a su simplicidad e interpretabilidad.
* **Random Forest (Bosque Aleatorio)**, un modelo más avanzado capaz de capturar relaciones complejas entre las variables.

La comparación entre ambos enfoques permitirá determinar cuál ofrece un mejor desempeño para predecir la cancelación de clientes y apoyar futuras estrategias de retención.

```python
# Librerías para entrenar los modelos con ambos métodos
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

**Modelo Base: Regresión Logística**

La Regresión Logística es uno de los algoritmos más utilizados en problemas de clasificación binaria. Su principal ventaja radica en la facilidad para interpretar los resultados y comprender cómo las distintas variables influyen en la probabilidad de abandono de un cliente.

En esta etapa se utilizará como punto de referencia para evaluar posteriormente modelos más complejos.

```python
# Modelo de Regresión Logística
lr_model = LogisticRegression(random_state = 0)
```

**Modelo Avanzado: Random Forest**

El algoritmo Random Forest construye múltiples árboles de decisión y combina sus predicciones para mejorar la capacidad de clasificación. Este enfoque suele capturar patrones más complejos en los datos y reducir el riesgo de sobreajuste.

Su desempeño será comparado con la Regresión Logística para identificar el modelo con mayor capacidad predictiva.

```python
# Modelo de Bosque Aleatorio
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)
```

```python
# Entrenar los modelos
lr_model.fit(X_train_st, y_train)
rf_model.fit(X_train_st, y_train)
```

```python
# Predicciones -> y_pred
lr_predictions = lr_model.predict(X_test_st)
rf_predictions = rf_model.predict(X_test_st)

lr_probabilities = lr_model.predict_proba(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)
```

### Evaluación del Desempeño de los Modelos
Una vez entrenados los modelos predictivos, es necesario evaluar su capacidad para identificar correctamente a los clientes con riesgo de abandono. Para ello se utilizarán métricas de clasificación que permitirán comparar el rendimiento de la Regresión Logística y Random Forest sobre datos no utilizados durante el entrenamiento.

Importante:
* __Exactitud__: Esta es la proporción de predicciones exactas entre todas las predicciones. Cuanto más cerca estemos del 100% de exactitud, mejor.
* __Precisión__: nos dice qué proporción de predicciones en la clase 1 es verdadero. En otras palabras, miramos la proporción de respuestas correctas solo en la clase objetivo.
* __Recall__: muestra el número de objetos de clase 1 reales que fuiste capaz de descubrir con tu modelo, por lo que aspira a minimizar los riesgos opuestos. Cuanto más cerca de 1, mejor.
* __F1 Score__: necesitamos una métrica armonizadora que tenga en cuenta el equilibrio entre las métricas

```python
# Librerías
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

from sklearn.metrics import confusion_matrix

from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
# Métricas de datos de validación -> Regresión Logística
print('Regresión Logística:')
print(f'- Exactitud: {accuracy_score(y_test, lr_predictions):.2f}')
print(f'- Precisión: {precision_score(y_test, lr_predictions):.2f}')
print(f'- Recall: {recall_score(y_test, lr_predictions):.2f}')
print(f'- F1 Score: {f1_score(y_test, lr_predictions):.2f}')

print()
cm_lr = confusion_matrix(y_test, lr_predictions)
print('Matriz de Confusión - Regresión Logística')
print(cm_lr)

print()
roc_auc_lr = roc_auc_score(y_test, lr_probabilities[:, 1])
print(f'ROC AUC Score - Regresión Logística: {roc_auc_lr:.2f}')
```

**Evaluación del Modelo de Regresión Logística**

Los resultados obtenidos muestran que la Regresión Logística presenta un desempeño sólido para la identificación de clientes con riesgo de abandono.
* Exactitud (92%): El modelo clasificó correctamente el 92% de los clientes evaluados, demostrando una alta capacidad de predicción general.
* Precisión (85%): Cuando el modelo identifica a un cliente como potencial abandono, acierta en el 85% de los casos, lo que reduce la probabilidad de invertir recursos de retención en clientes que realmente no abandonarían el servicio.
* Recall (83%): El modelo logró detectar el 83% de los clientes que efectivamente cancelaron su membresía, permitiendo anticipar una gran proporción de los abandonos futuros.
* F1 Score (84%): El equilibrio entre precisión y recall indica que el modelo mantiene un buen rendimiento tanto para identificar clientes en riesgo como para minimizar falsas alarmas.
* ROC AUC (97%): El modelo presenta una capacidad excelente para diferenciar entre clientes que permanecerán activos y aquellos que cancelarán su membresía, lo que demuestra un alto potencial para apoyar estrategias preventivas de retención.

En conjunto, estos resultados sugieren que la Regresión Logística constituye una herramienta confiable para identificar clientes con riesgo de abandono y apoyar la toma de decisiones orientadas a mejorar la fidelización de usuarios en Model Fitness.

Además, la Matriz de Confusión entregó los siguientes resultados, recordando que permanencia es churn = 0 y abandono es churn = 1:
* Verdaderos Negativos: 574
* Verdaderos Positivos: 164
* Falsos Positivos: 28
* Falsos Negativos: 34
La Regresión Logística logró clasificar correctamente la gran mayoría de los clientes, identificando 574 clientes que permanecieron activos y 164 clientes que cancelaron su membresía.

Sin embargo, el modelo también cometió algunos errores de clasificación. Se registraron 28 falsos positivos, correspondientes a clientes que fueron identificados como potenciales abandonos, pero que finalmente permanecieron en el gimnasio. Asimismo, se detectaron 34 falsos negativos, es decir, clientes que terminaron cancelando su membresía sin que el modelo lograra anticiparlo.

Desde una perspectiva de negocio, los falsos negativos representan el riesgo más relevante, ya que corresponden a clientes perdidos sobre los cuales no se aplicaron acciones preventivas de retención. No obstante, la cantidad observada es relativamente baja considerando el tamaño total de la muestra, lo que refleja una buena capacidad predictiva del modelo.

```python
# Métricas de datos de validación -> Bosque Aleatorio
print('Bosque Aleatorio:')
print(f'- Exactitud: {accuracy_score(y_test, rf_predictions):.2f}')
print(f'- Precisión: {precision_score(y_test, rf_predictions):.2f}')
print(f'- Recall: {recall_score(y_test, rf_predictions):.2f}')
print(f'- F1 Score: {f1_score(y_test, rf_predictions):.2f}')

print()
cm_rf = confusion_matrix(y_test, rf_predictions)
print('Matriz de Confusión - Bosque Aleatorio')
print(cm_rf)

print()
roc_auc_rf = roc_auc_score(y_test, rf_probabilities[:, 1])
print(f'ROC AUC Score - Bosque Aleatorio: {roc_auc_rf:.2f}')
```

Sin embargo, al analizar métricas directamente relacionadas con la detección de clientes en riesgo, la Regresión Logística obtuvo resultados ligeramente superiores, alcanzando una precisión del 85%, un recall del 83% y un F1 Score del 84%, frente al 84%, 81% y 83% obtenidos por el Bosque Aleatorio.

La matriz de confusión refuerza esta conclusión. La Regresión Logística logró identificar correctamente 164 clientes que efectivamente cancelaron su membresía, mientras que el Bosque Aleatorio identificó 161 clientes. Además, presentó una menor cantidad de falsos negativos (34 frente a 37), lo que resulta especialmente relevante desde una perspectiva de negocio, ya que cada falso negativo representa un cliente que abandona el gimnasio sin ser detectado previamente por el modelo.

Por lo tanto, considerando la capacidad de identificar clientes con riesgo de abandono y minimizar pérdidas potenciales, la Regresión Logística fue seleccionada como el modelo de mejor desempeño para este proyecto.

## Segmentación Estratégica de Clientes mediante Clustering
Hasta este punto del análisis se ha identificado qué factores están relacionados con la cancelación de membresías y se ha desarrollado un modelo predictivo capaz de anticipar el abandono de clientes.

Sin embargo, no todos los usuarios presentan los mismos hábitos, necesidades o niveles de compromiso con el gimnasio. Por esta razón, resulta relevante complementar el análisis mediante técnicas de segmentación que permitan identificar grupos de clientes con características similares.

Para ello, se aplicará el algoritmo **K-Means**, una técnica de Machine Learning no supervisado que agrupa observaciones según patrones de comportamiento compartidos. A diferencia de los modelos predictivos, este enfoque no utiliza la variable de cancelación como referencia, sino que busca descubrir estructuras ocultas dentro de los datos.

La identificación de estos segmentos permitirá responder preguntas clave como:
* ¿Existen perfiles claramente diferenciados de clientes?
* ¿Qué características definen a cada grupo?
* ¿Qué segmentos presentan mayor riesgo de abandono?
* ¿Cómo pueden adaptarse las estrategias de fidelización según el perfil de cada cliente?

Los resultados obtenidos permitirán diseñar acciones más específicas y efectivas para mejorar la experiencia de los usuarios y fortalecer la retención en Model Fitness.

### Preparación de los Datos para la Segmentación
Como el objetivo del clustering es descubrir grupos de clientes de forma autónoma, se excluye la variable churn, ya que representa información conocida sobre la cancelación y podría sesgar la formación de los clústeres.

Asimismo, las variables serán estandarizadas para garantizar que todas contribuyan de manera equilibrada al algoritmo de agrupamiento.

#### Estandariza los datos.
Antes de aplicar técnicas de agrupamiento, se estandarizaron las variables del conjunto de datos. La estandarización permite que todas las variables tengan la misma importancia durante la formación de los grupos y evita sesgos en el análisis.

```python
# Librería
scaler = StandardScaler()
```

```python
# Eliminar la columna 'Churn' -> Variable Objetivo
X = gym.drop('churn', axis = 1)
```

```python
# Estandarización previo a pasarlo a algoritmo
sc = StandardScaler()
X_sc = sc.fit_transform(X)
```

```python
# Convertir a DataFrame
X_sc = pd.DataFrame(
    X_sc,
    columns = X.columns
)
```

```python
# Visualizar
print(X_sc.head(10))
```

### Determinación del Número Óptimo de Clústeres
Antes de aplicar el algoritmo K-Means, es necesario estimar cuántos grupos naturales existen dentro de la base de clientes. Una elección adecuada del número de clústeres permite obtener segmentos más homogéneos y representativos de los distintos perfiles de usuarios.

Para ello, se utilizará un análisis jerárquico basado en una matriz de distancias construida a partir de las variables estandarizadas. Este enfoque permite visualizar las similitudes entre los clientes mediante un dendrograma, facilitando la identificación de agrupaciones naturales dentro de los datos.

El objetivo de esta etapa es determinar una cantidad razonable de segmentos antes de aplicar el algoritmo de clustering definitivo.

```python
# Librería
from scipy.cluster.hierarchy import dendrogram, linkage
```

#### Construcción de la Matriz de Distancias

Con el fin de medir el nivel de similitud entre los clientes, se calculará una matriz de distancias utilizando las variables previamente estandarizadas. Esta matriz servirá como base para el agrupamiento jerárquico y permitirá visualizar cómo los distintos usuarios se relacionan entre sí.

```python
# Almacenamiento de la tabla con los objetos vinculados -> Matriz de distancias
linked = linkage(
    X_sc,
    method = 'ward'
)
```

#### Visualización de Agrupaciones Potenciales mediante Dendrograma
El dendrograma representa gráficamente el proceso de agrupamiento jerárquico de los clientes. La altura de las uniones refleja la distancia entre grupos, permitiendo identificar posibles cortes naturales que sugieren la cantidad óptima de clústeres presentes en los datos.

```python
# Creación de Gráfico
plt.figure(figsize = (6,6))

dendrogram(
    linked,
    orientation = 'top'
)

plt.title('Dendrograma Clientes')
plt.show()
```

#### Visualización de Agrupaciones Potenciales mediante Dendrograma
El dendrograma permite visualizar cómo los clientes se agrupan según sus similitudes en las variables analizadas. Cada unión representa la combinación de grupos de clientes con características cada vez más parecidas, mientras que la altura de las ramas refleja la distancia entre dichas agrupaciones.

Al analizar la estructura del gráfico, se observan saltos significativos en las distancias de agrupamiento, lo que sugiere la existencia de varios segmentos claramente diferenciados dentro de la base de clientes.

De manera visual, el dendrograma indica que una solución entre 4 y 5 clústeres podría representar adecuadamente la estructura de los datos. Para mantener un nivel adecuado de segmentación y obtener perfiles de clientes más específicos, se trabajará con 5 clústeres en la siguiente etapa del análisis mediante el algoritmo K-Means.

### Segmentación de Clientes mediante K-Means
Una vez identificado el número óptimo de grupos mediante el análisis jerárquico y el dendrograma, el siguiente paso consiste en aplicar el algoritmo K-Means para clasificar a los clientes en segmentos con características similares.

A diferencia de los modelos predictivos desarrollados anteriormente, este enfoque no busca anticipar la cancelación de clientes, sino descubrir patrones de comportamiento que permitan comprender mejor la composición de la base de usuarios.

La segmentación permitirá identificar distintos perfiles de clientes según variables como frecuencia de asistencia, antigüedad, duración de contratos, gasto adicional y participación en actividades grupales. Esta información resulta especialmente valiosa para diseñar estrategias de retención y fidelización adaptadas a las necesidades de cada segmento.

Considerando los resultados obtenidos en el dendrograma, se trabajará con cinco clústeres, permitiendo capturar las diferencias más relevantes entre los distintos perfiles de clientes de Model Fitness.

```python
# Librería
from sklearn.cluster import KMeans
```

#### Configuración del Modelo de Clustering
Con base en el análisis previo, se configura el algoritmo K-Means para generar cinco grupos de clientes. Cada observación será asignada al clúster cuyo centroide represente de mejor manera sus características y comportamiento.

```python
# Establecer el modelo y número de clústeres
km_model = KMeans(
    n_clusters = 5,
    random_state = 0
)
```

#### Asignación de Clientes a Segmentos
El algoritmo analiza las similitudes entre los clientes y los agrupa automáticamente en segmentos homogéneos. Como resultado, cada usuario recibe una etiqueta que indica el clúster al que pertenece.

```python
# Aplicar el algoritmo a los datos y formar un vector de clúster
labels = km_model.fit_predict(X_sc)
```

```python
# Almacena etiquetas de clúster en el campo de nuestro cojunto de datos
gym['cluster'] = labels
```

#### Distribución Inicial de Clientes por Clúster
Una vez generados los segmentos, resulta relevante conocer cuántos clientes forman parte de cada grupo. Esto permitirá evaluar el tamaño relativo de los clústeres y comprender mejor la estructura general de la base de clientes.

```python
# ¿Cuántos clientes hay por grupo?
km_users_group = gym['cluster'].value_counts().sort_index()
print(km_users_group)
```

#### Primeras Observaciones sobre la Segmentación
La aplicación del algoritmo K-Means permitió identificar cinco grupos diferenciados de clientes dentro de la cadena Model Fitness. La distribución de usuarios entre los clústeres muestra que ningún segmento concentra de forma excesiva la totalidad de la base de clientes, lo que sugiere una segmentación relativamente equilibrada.

El Clúster 0 es el más numeroso, agrupando aproximadamente el 28% de los clientes analizados, mientras que el Clúster 4 representa el segmento más pequeño con cerca del 14% de los usuarios.

Esta distribución indica que el algoritmo logró identificar perfiles de clientes con tamaños suficientes para realizar análisis posteriores y extraer conclusiones de negocio relevantes. En las siguientes etapas se estudiarán las características promedio de cada grupo con el fin de comprender sus patrones de comportamiento, nivel de compromiso con el gimnasio y posible riesgo de cancelación.

#### Caracterización de los Segmentos de Clientes
Una vez definidos los cinco clústeres mediante K-Means, el siguiente paso consiste en analizar las características promedio de cada grupo. Este análisis permite comprender qué distingue a cada segmento en términos de antigüedad, frecuencia de asistencia, tipo de contrato, gasto adicional y nivel de compromiso con el gimnasio.

El objetivo es identificar perfiles de clientes con comportamientos similares para detectar oportunidades de fidelización, anticipar riesgos de cancelación y diseñar estrategias específicas para cada segmento.

```python
# Valores medios de Características para clústeres
km_cluster_mean = gym.groupby('cluster').mean().round(2)
print(km_cluster_mean.T)
```

Particularmente, el Clúster 2 registra la menor tasa de abandono (2%), destacando además por una elevada participación en programas de recomendación y contratos cercanos al año completo.

Por el contrario, los clústeres 0 y 4 presentan los mayores niveles de cancelación, alcanzando tasas del 51% y 45%, respectivamente. En el caso del Clúster 4, además, destaca la ausencia de proximidad geográfica al gimnasio, lo que podría estar influyendo negativamente en la retención de clientes.

### Visualización de la Separación entre Clústeres
Una vez identificados los segmentos de clientes mediante el algoritmo K-Means, resulta fundamental visualizar cómo se distribuyen los distintos grupos en función de las variables más relevantes del estudio.

A continuación, se construye una función que permitirá representar gráficamente diferentes combinaciones de variables para analizar la distribución de los clientes dentro de cada clúster.

```python
# Función para la Visualización de Clústeres
def show_clusters_on_plot(df, x_axis, y_axis):
    plt.figure(figsize = (6,6))

    sns.scatterplot(
        data = df,
        x = x_axis,
        y = y_axis,
        hue = 'cluster',
        palette = 'Set1'
    )

    plt.title(f'{x_axis} vs {y_axis}')
    plt.show()
```
